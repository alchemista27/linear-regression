```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm


# Load data
raw_data = pd.read_csv('Salary_dataset.csv', index_col = 'Unnamed: 0')
raw_data.head()


# Checking the missing value
raw_data.info()


# Creating scatter plot
fig = plt.figure(figsize=(14,7))
sns.scatterplot(data= raw_data, x='YearsExperience', y='Salary')
plt.title('Years of Experience vs Salary', fontsize=20, pad=20)
plt.xlabel('Years of Experience', fontsize=16)
plt.ylabel('Salary', fontsize=16)
plt.show()
```

```python
# Instantiate the regression model
model = sm.OLS(raw_data['YearsExperience'], raw_data['Salary'])
res = model.fit()

res.summary()
```
